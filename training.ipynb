```
## ARIMA model: 

## ARIMA(p,d,q)

# Auto-Regressive (p) -> Number of autoregressive terms.

# Integrated (d) -> Number of nonseasonal differences needed for stationarity.

# Moving Average (q) -> Number of lagged forecast errors in the prediction equation.
```

```
import pmdarima as pmd
```

```
# arima_model = pmd.auto_arima(train, 
#                          start_p=0, d=1, start_q=0,
#                          max_p = 5, max_d =5, max_q =5,
#                          start_P=0, D=1,start_Q=0,
#                          max_P=5, max_D =5,max_Q=5,
#                          m=365,
#                          seasonal= True,
#                          error_action='warn',
#                          trace=True,
#                          surpress_warnings = True,
#                          stepwise=True,
#                          random_state=20,
#                          n_fits=50)
```

```
arima_model = pmd.auto_arima(train, 
                             start_p=0, d=1, start_q=0,
                             m=7,
                             seasonal = True,
                             test="adf", 
                             supress_warnings = True,
                             trace=True)
```

```
arima_model.summary()
```

```
conf_int  = arima_model.predict(100,return_conf_int=True,alpha=0.05)
print(conf_int)
```

```
prediction = pd.DataFrame(arima_model.predict(n_periods = 128),index=test.index)

prediction.columns = ['PREDICTED_CARS']

prediction
```

```
conf_df = pd.concat(prediction,conf_int], axis = 1)
```
