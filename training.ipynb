```
## using np.where 

## np.where(condition, value when condition is true, value when condition is false)
## think of np.where() as an if statement

df['RANDOM'] = np.random.randint(0,2, size=len(df)) ## creates a random column of numbers 1,2 and 3## create a column of random 0 and 1 

df['RANDOM'] = np.where(df['RANDOM'] == 0, 'Y', 'N') ## if value = 0 replace with Y, else replace with N

df.head()
```

```
## create a weather column taking three values: Sun, Rain, Dry

df['WEATHER'] = np.random.randint(1,4, size=len(df)) ## creates a random column of numbers 1,2,3

## np.where(condition, value when condition is true, value when condition is false)
## think of np.where() as an if statement


## create conditions list 
condition = [
             df['WEATHER'] == 1,
             df['WEATHER'] == 2,
             df['WEATHER'] == 3
             ]

#create output list for conditions
output = [ 'Sun', 'Rain', 'Dry']


df['WEATHER'] = np.select(condition, output, default=np.nan) ##updates the column to reflect the condition and output list, returns NaN if conditions are not met
```

**GROUP BY**

Process of splitting, applying function and combining results

```
month_grouped = df.groupby(['MONTH_2'])

month_grouped.get_group('January') ## just shows data for January for all years 

month_grouped.get_group('January').value_counts('WEATHER') ##how many times does Sun, Rain and Dry appear in month of January

## Question 

## a) using .loc function, extract data from only January and see if it ties up with the above

## b) Find the mean for all columns in year 2021

year_grouped = df.groupby(['YEAR'])

year_grouped.get_group(2021).mean()
```

```
## getting mean, median and mode grouped by Year

year_grouped = df.groupby(['YEAR'])


year_grouped.agg(['mean', 'median'])

## Question

## a) Using the above function, try to extract data for London_Cars only. Hint: using .loc function 

## b) Add in the mode into the above list. Why do you think this errors?

## c) Display the sum of columns by year - ie total number of cars in london by year, total number of motorbikes by year ect 

## d) HARD: group by year and month and work out sum of columns. Answer in below cell
```

**PLOTTING USING MATPLOTLIB**

https://matplotlib.org/

```
## import packages

import matplotlib.pyplot as plt ##matplotlib package
import seaborn as sns ## seaborn 
import plotly.express as px ## ploty package - interactive graphs
%matplotlib inline 
```

```
## heatmap of missing values

sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis') 
```

```
plt.figure(figsize=(25,10)) ## width x height control 

plt.style.use('seaborn') ## run command plt.style.available to see what is available

## plot(x,y)

plt.plot(df['DATE'], df['LONDON_CARS'], linewidth=4)

plt.xlabel('Date', fontweight='bold')
plt.ylabel('no. cars', fontweight='bold')
plt.title('Number of cars in London from 1 March 2020 to 8 May 2022', fontweight='bold')
plt.show()
```

```
##plotly plot

fig = px.line(df, x='DATE', y=['LONDON_CARS'])

fig.update_xaxes(title_text='Date')
fig.update_yaxes(title_text='no. of cars')

fig.show()
```

```
plt.figure(figsize=(25,10)) ## width x height control 

plt.style.use('ggplot') ## run command plt.style.available to see what is available

## plot(x,y)

plt.plot(df['DATE'], df['LONDON_CARS'], linewidth=4)
plt.plot(df['DATE'], df['NORTH_EAST_CARS'])
plt.plot(df['DATE'], df['NORTHERN_IRELAND_CARS'])
plt.plot(df['DATE'], df['SOUTHEND_CARS'])
plt.plot(df['DATE'], df['GREATER_MANCHESTER_CARS'])
plt.plot(df['DATE'], df['READING_CARS'])

plt.xlabel('Date', fontweight='bold')
plt.ylabel('no. cars', fontweight='bold')
plt.title('Number of Cars by Region', fontweight='bold')

plt.legend(['London', 'North East', 'Northen Ireland', 'Southend', 'Greater Manchester', 'Reading'])

plt.show()
```
